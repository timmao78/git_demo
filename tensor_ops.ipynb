```python
import torch
```

```python
my_torch = torch.arange(10)
my_torch
```

```python
#reshape and view
my_torch1 = my_torch.reshape(2,5)
my_torch1
```

```python
my_torch
```

```python
my_torch2 = my_torch.view(2,5)
my_torch2
```

```python
#with reshape and view, they will update
my_torch[1] = 4141
my_torch
```

```python
my_torch1
```

```python
my_torch2
```

```python
#slices
my_torch3 = my_torch.reshape(5,2)
my_torch3
```

```python
my_torch3[:,1]
```

```python
my_torch3[0,:]
```

```python
my_torch3[:,1:]
```
